## 필요한_데이터 = ['식품코드', '식품명', '식품대분류명', '에너지(kcal)', '단백질(g)', '지방(g)', '탄수화물(g)', '식품중량']

```python
import pandas as pd
```

```python
# 처리할 JSON 파일 경로
file_path = 'D:/student/LHT/works/crawling/food_data.json'

# JSON 파일 읽기
df = pd.read_json(file_path)

# 데이터 프레임의 상위 5개 행 확인
df.head()
```

```python
# 데이터 프레임의 정보 확인 (열의 데이터 타입 등)
df.info()
```

```python
# 차례대로 '식품코드', '식품명', '식품대분류명', '에너지(kcal)', '단백질(g)', '지방(g)', '탄수화물(g)', '식품중량' 열만 선택
df_preprocessed = df[['foodCd', 'foodNm', 'foodLv3Nm', 'enerc', 'prot', 'fatce', 'chocdf', 'foodSize']]
df_preprocessed
```

```python
# 처리한 데이터프레임을 JSON 파일로 저장. 저장 옵션 : 각 행을 JSON 객체(키 - 값 구조)로 저장, 아스키 코드 강제 변환 금지, 들여쓰기 4
df_preprocessed.to_json('D:/student/LHT/works/crawling/preprocessed_food_data.json', orient='records', force_ascii=False, indent=4)
```
